# Life Expectancy and GDP

## Introduction

This project aims to explore the relationship between a country's economic output (GDP) and life expectancy, utilizing data from the World Health Organization. The analysis seeks to uncover insights into how economic factors may influence public health outcomes across six nations.

### Questions to Explore
1. Trends Over Time:
   * How has life expectancy evolved over the years in the selected countries?
   * What are the trends in GDP growth from 2000 to 2015 for these nations?
2. Correlation Analysis
   * Is there a noticeable correlation between a country's GDP and its life expectancy?
   * How strong is this correlation across different countries?
3. Comparative Analysis
   * What is the average life expectancy across these nations?
   * How does life expectancy vary within and across countries?
4. Influence of Economic Growth:
   * Which economic factors, such as GDP growth rates, are associated with the highest increases in life expectancy?
5. Regional Comparisons:
   * Are there significant differences in life expectancy trends between developed and developing nations in the dataset?

### Exploratory Data Analysis (EDA)
To explore the distributions of GDP and life expectancy, we will use various visualizations to identify patterns and outliers in the data.

#### Distribution of GDP
To understand the distribution of GDP values, we first create a histogram. The histogram helps to visualize the frequency distribution of GDP values across the six countries.

**Observation**

From the histogram, we observe that the distribution of GDP values is right-skewed. This can mean that most of the countries in our dataset have GDP values clustered on the lower end, with fewer countries having very high GDP values. 

This right skewness indicates the presence of outliers, which are countries with significantly higher GDPs. The skewness could be due to different economic factors.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df.GDP, kde=False, bins=30)
plt.title("Distribution of GDP")
plt.xlabel("GDP in Trillions of U.S. Dollars")
plt.ylabel("Frequency")
plt.show()
```

#### Distribution of Life Expectancy (LEABY)

We also examine the distribution of life expectancy values using box plot to identify patterns and outliers.

**Observation**

From the box plot we can observe the median GDP value, which is represented by the line within the box, indicating the central tendency of the data. This shows that the distribution of life expetancy values 
are clustered on the higher end. 

Even though most countries are on the higher end of LEABY, there are several outliers which is indicated by the trailing small dots towards the bottom of the box plot.

```python
plt.figure(figsize=(10,6))
sns.boxplot(y=df.LEABY)
plt.title("Box Plot of Life Expectancy at Birth (years)")
plt.ylabel("Life Expectancy at Birth (years)")
plt.show()
```

#### Relationship Between GDP and Life Expectancy

To explore the relationship between GDP and life expectancy, we use scatter plots. These plots help us identify any correlation between a country's economic output and the health of its citizens.

**Observation**

From the scatter plots, we can observe the following:

- In general, there seems to be a positive correlation between GDP and life expectancy. As GDP increases, life expectancy tends to increase as well.
- The United States and China show particularly strong positive correlations between GDP and life expectancy, with higher GDP values corresponding to higher life expectancy.
- Other countries, such as Zimbabwe, show a more modest increase in life expectancy with increasing GDP, possibly due to different stages of economic development or other socio-economic factors.
- The scatter plots also reveal some outliers and variations within each country, indicating that while GDP is an important factor, it is not the sole determinant of life expectancy.

These scatter plots help to visually confirm the relationship between economic prosperity and health outcomes across different countries.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='LEABY', y='GDP', hue='Country')
plt.title('GDP vs Life Expectancy by Country')
plt.xlabel('GDP in Trillions of U.S. Dollars')
plt.ylabel('Life Expectancy at Birth (years)')
plt.show()
```

### Statistical Analysis
In this section, we calculate basic statistics such as mean, median, and standard deviation for GDP and life expectancy. 

#### Descriptive Statistics
We begin by calculating descriptive statistics for GDP and life expectancy

```python
# Calculate descriptive stats for GDP
gdp_mean = np.mean(df.GDP)
gdp_median = np.mean(df.GDP)
gdp_std = np.std(df.GDP)

print('GDP')
print('\tGDP Mean:', gdp_mean)
print('\tGDP Median:', gdp_median)
print('\tGDP Standard Deviation:', gdp_std)
print('\n')

# Calculate descriptive stats for LEABY
leaby_mean = np.mean(df.LEABY)
leaby_median = np.median(df.LEABY)
leaby_std = np.std(df.LEABY)

print('Life Expectancy at Birth (years)')
print('\tLife Expectancy Mean:', leaby_mean)
print('\tLife Expectancy Median:', leaby_median)
print('\tLife Expectancy Standard Deviation:', leaby_std)
```

#### Correlation Analysis
To understand the relationship between GPD and life expectancy, we calculate the Pearson correlation coefficient. This measures the strength and direction of the relationship between two variables.

A positive correlation coefficient indicates a positive relationship between GDP and Life Expectancy, meaning as GDP increases, life expectancy tends to increase as well. 

In our analysis, a positive correlation coefficient suggests that higher economic output(GDP) can be associated with higher life expectancy in the six countries studied.

```python
# Calculate Pearson correlation coefficient between GDP and LEABY
corr, pval = pearsonr(df.GDP, df.LEABY)
print("Pearson Correlation Coefficient between GDP and Life Expectancy:",corr)
```
